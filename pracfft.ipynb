```python
%matplotlib inline
import math as mat
import numpy as np
import matplotlib.pyplot as pl
import scipy.fftpack as pf
```

###### Leemos os dados das manchas Solares anuais

```python
dados1 = np.genfromtxt('monthssn.dat')
print dados1
x1 =dados1[:,1]
y1 =dados1[:,3]
```

###### Plotamos  essos dados

```python
pl.figure(figsize=(10,7))
pl.plot(x1,y1,'-r')
pl.xlim([1748, 2015])
pl.title('International sunspot number $R_i$ (pure)')
pl.xlabel('time(year)')
pl.ylabel('Sunspot number R_i')
```

###### Agora  aplicamos a  tranformda de  Fourier

```python
fhat = pf.fft(y1)
```

###### plotamos  os coeficientes de Fourier no  plano compleixo

```python
pl.figure(figsize=(10,7))
pl.scatter(fhat.real,fhat.imag)
pl.title('Fourier Coefficients')
pl.xlabel('Real Axis')
pl.ylabel('Imaginary Axis')
pl.show()
```

###### Fourier Transform of a simple signal

```python
pl.figure(figsize=(10,7))
gap = x1[1]-x1[0]
tam = len(x1)
omegas = pf.fftfreq(tam,gap)
ampfft = abs(fhat)*2./tam
pl.plot(omegas,ampfft, '-')
pl.xlabel('frequency')
pl.ylabel('Fourier Transform amplitude')
pl.xlim([-1.5,1.5])
pl.ylim([0,30])
pl.show()
```

###### Convertimos  as frequencias em  anos

```python
pl.figure(figsize=(10,7))
period = 1./omegas
pl.plot(period,ampfft, '-')
pl.xlim([0.1, 40])
pl.ylim([0,35])
pl.xlabel('Period [year]')
pl.ylabel('Fourier Transform amplitude')
mx = period[1:158]
my = ampfft[1:158]
max_y = max(my)
max_x = mx[my.argmax()]
pl.text(max_x, max_y, str((max_x)))
pl.show()
```
