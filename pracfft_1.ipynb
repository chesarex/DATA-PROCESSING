```python
%matplotlib inline
import math as mat
import numpy as np
import scipy
import matplotlib.pyplot as pl
from matplotlib.pylab import hist
import scipy.fftpack as pf
```

###### Leemos os dados das manchas Solares mensuais

```python
dados1 = np.genfromtxt('monthssn.dat')
x1 =dados1[:,1]
y1 =dados1[:,2]
f =dados1[:,3]
```

```python
pl.figure(1)
pl.plot(x1,y1,'-r')
pl.xlim([1748, 2015])
pl.title('International sunspot number $R_i$ (pure)')
pl.xlabel('time(year)')
pl.ylabel('Sunspot number R_i')
pl.show()
```

###### Agora  aplicamos a  tranformda de  Fourier e  plotamos  os coeficientes de Fourier

```python
fnoisyhat = pf.fft(y1)
pl.scatter(fnoisyhat.real,fnoisyhat.imag)
pl.title('Fourier Coefficients')
pl.xlabel('Real Axis')
pl.ylabel('Imaginary Axis')
#pl.xlim([-25000,190000])
pl.show()
```

###### Espectro de  Frequencias

```python
gap = x1[1]-x1[0]
tam = len(x1)
omegas = pf.fftfreq(tam,gap)
ampfftnoisy = abs(fnoisyhat)*2./tam
pl.plot(omegas,ampfftnoisy, '-')
pl.xlabel('frequency')
pl.ylabel('Fourier Transform amplitude')
pl.xlim([-1.5,1.5])
pl.ylim([0,35])
pl.show()
```

###### Convertimos  as frequencias em  anos

```python
period = 1./omegas
pl.plot(period, ampfftnoisy, '-')
pl.xlabel('Period [year]')
pl.ylabel('Fourier Transform amplitude')
mx = period[1:158]
my = ampfftnoisy[1:158]
max_y = max(my)
max_x = mx[my.argmax()]
pl.text(max_x, max_y, str((max_x)))
pl.xlim([0.1, 40])
pl.ylim([0, 35])
pl.show()
```

```python
pl.subplot(2,2,1)
threshold_freq = 0.7
fnoisyhat[abs(omegas)>threshold_freq] = 0
fclean = pf.ifft(fnoisyhat)
pl.plot(x1,y1,'-r')
pl.xlim([1748, 2015])
pl.xlabel('time')
pl.ylabel('signal')
pl.title('International sunspot number $R_i$ (pure)')
pl.subplot(2,2,2)
pl.plot(x1,fclean,'-b')
pl.title('International sunspot number $R_i$ (pure)')
pl.xlabel('time')
pl.ylabel('Clean signal')
pl.xlim([1748, 2015])
pl.subplot(2,2,3)#plot fft clean
fclean_noisy = pf.fft(fclean)
ampffclean = abs(fclean_noisy)*2./tam
pl.plot(omegas,ampffclean , '-')
pl.xlabel('frequency')
pl.ylabel('Fourier Transform amplitude (clean noisy)')
pl.xlim([-1.5,1.5])
pl.ylim([0,35])
#singnal-signal(noisy)
pl.subplot(2,2,4)
pl.plot(x1,y1-fclean)
pl.xlabel('time')
pl.ylabel('Difference')
pl.xlim([1748, 2015])
pl.show()
```

```python
pl.plot(x1,y1,'-r',x1,fclean)
pl.xlim([1748, 2015])
pl.show()
```

###### Comparamos  com os dados originais de  manchas solares anuais

```python
pl.plot(x1,f,'-r',x1,fclean,'-b')
pl.xlim([1748, 2015])
pl.show()
```

###### # Cleaning by rejecting the lowest coefficients: new signal

```python
pl.figure(3)
pl.subplot(2,2,1)
pl.plot(x1,y1,'-r'), pl.xlim([1748, 2015])
pl.title('International sunspot number $R_i$ (pure)')
pl.subplot(2,2,3)
fnoisyhat_norm = abs(fnoisyhat)*2./tam
M = scipy.mean(fnoisyhat_norm)
sigma = scipy.std(fnoisyhat_norm)
aux = scipy.ones((len(omegas),))
pl.plot(omegas,ampfftnoisy, '-')
pl.xlabel('frequency')
pl.ylabel('Fourier Transform amplitude')
pl.xlim([-1.5,1.5])
pl.ylim([-3,35])
pl.plot(omegas,aux*M,'r')
pl.plot(omegas,aux*M-sigma,'r--')
pl.plot(omegas,aux*M+sigma,'r--')
fnoisyhat[abs(fnoisyhat_norm-M)<sigma] = 0
fclean = pf.ifft(fnoisyhat)
pl.subplot(2,2,2)
pl.plot(x1,fclean, '-g')
pl.xlabel('time')
pl.ylabel('Clean signal')
pl.xlim([1748, 2015])
pl.subplot(2,2,4)
pl.plot(x1,f-fclean)
pl.xlim([1748, 2015])
pl.xlabel('time')
pl.ylabel('Difference')
pl.show()
```

```python
pl.plot(x1,y1,'-r',x1,fclean,'-b')
pl.xlim([1748, 2015])
pl.show()
```

```python
pl.plot(x1,f,'-r',x1,fclean,'-b')
pl.xlim([1748, 2015])
pl.show()
```
